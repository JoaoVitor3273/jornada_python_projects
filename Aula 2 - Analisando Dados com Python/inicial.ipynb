### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Base de dados e arquivos: https://drive.google.com/drive/folders/1uDesZePdkhiraJmiyeZ-w5tfc8XsNYFZ?usp=drive_link

```python
#passo a passo do projeto
#pip install openpyxl ipykernel nbformaf plotly
#keagle = site para achar base de dados

#passo 1: importar a base de dados
import pandas as pd

tabela = pd.read_csv("cancelamentos.csv")


#passo 2: visualizar a base de dados
    #entender a basse de dados
    #possiveis problemas na base
    #descartando colunas inuteis
tabela = tabela .drop(columns="CustomerID")
display(tabela)
```

```python
#passo 3: corrigir problemas da base de dados (tratamentos de dados)
#valores em formatos errados - OK
#inormações vazias

display(tabela.info())
tabela = tabela.dropna()
display(tabela.info())
```

```python
#passo 4: analise inicial (entender os cancelamentos)

#quantidade
display(tabela["cancelou"].value_counts())

#porcentagem
display(tabela["cancelou"].value_counts(normalize=True))
```

```python
#passo 5: analise detalhada (causa dos cancelamentos)
import plotly.express as px

#cria graico de uma coluna
grafico =px.histogram(tabela, x= "duracao_contrato", color="cancelou")

#mostra o grafico
grafico.show()

#cria grafico para todas as colunas da base de dados
for coluna in tabela.columns:
    #cria graico de uma coluna
    grafico = px.histogram(tabela, x=coluna, color="cancelou")
    #mostra o grafico
    grafico.show()

```

```python
#todos os clientes do contrato mensal cancelaram
#todos os clientes que ligaram mais de 4 vezes cancelaram suas assinaturas
#todos os clientes que atrasam mais de 20 dias no pagamento cancelaram suas assinaturas    

#cancelemntos 56%

#iltrar a base de dados
#duracao_contrato -> diferente de monthly
condicao = tabela["duracao_contrato"] != "Monthly"
tabela = tabela[condicao]

#ligacoes_callcenter -> menor ou igual a 4
condicao = tabela["ligacoes_callcenter"] <= 4
tabela= tabela[condicao]

#dias_atraso -> menos ou igual a 20
condicao = tabela["dias_atraso"] <= 20
tabela = tabela[condicao]

#numero de cancelamentos
display(tabela["cancelou"].value_counts())
#em %
display(tabela["cancelou"].value_counts(normalize=True))
```
